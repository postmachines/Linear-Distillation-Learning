```python
import pandas as pd
import numpy as np
pd.set_option('max_columns', 50)
```

```python
def make_tables():
    df_dict = {}
    for shot in [10, 50, 100, 200, 300]:
        df_dict[shot] = []
    for ds_name in ['customer', 'fashion_mnist', 'svhn']:
        print(ds_name)
        df0 = pd.read_csv(f'results/02-02-2021/results_{ds_name}_svd_o2m.csv')
        df = pd.read_csv(f'results/02-02-2021/results_{ds_name}_svd_o2m_v2.csv')
        df1 = pd.read_csv(f'results/02-02-2021/results_{ds_name}_bidir.csv')
        df2 = pd.read_csv(f'results/02-02-2021/results_{ds_name}_o2m.csv')
        df3 = pd.read_csv(f'results/02-02-2021/results_{ds_name}_logreg.csv')
        df4 = pd.read_csv(f'results/02-02-2021/results_{ds_name}_tree.csv')
        df5 = pd.read_csv(f'results/02-02-2021/results_{ds_name}_mlp.csv')
        df6 = pd.read_csv(f'results/02-02-2021/results_{ds_name}_o2mlog.csv')
        df1['model'] = 'Bidirectional distillation'
        df['model'] = 'SVD full O2M distillation'
        df0['model'] = 'SVD simple O2M distillation'
        df2['model'] = 'O2M distillation'
        df3['model'] = 'Logistic Regression'
        df4['model'] = 'Decision Tree'
        df5['model'] = 'MLP'
        df6['model'] = 'Logdistill O2M'
        df_concat = pd.DataFrame(columns=['model', 'dataset', 'train_shot', 'accuracy'])
        for single_df in [df0, df, df1, df2, df3, df4, df5, df6]:
            df_concat = pd.concat([df_concat, (single_df[single_df['train_shot'].isin([10, 50, 100, 200, 300])]
             .groupby(by=['model', 'dataset', 'train_shot'])
             .agg({'accuracy': 'max'})
             .reset_index())])
        for shot in [10, 50, 100, 200, 300]:
            df_dict[shot].append(df_concat[df_concat['train_shot'] == shot])
    for shot in [10, 50, 100, 200, 300]:
        df_dict[shot] = pd.concat(df_dict[shot]).drop(columns=['train_shot']).sort_values(by=['dataset', 'accuracy'])
    return df_dict
```

```python
df_dict = make_tables()
```

```python
df_logMN = pd.read_csv('results/02-02-2021/results_mnist_o2mlog.csv')
```

```python
for shot in [10, 50, 100, 200, 300]:
    (df_dict[shot]
     .pivot(index='model', columns='dataset', values='accuracy')
     .reset_index()
     .to_csv(f'results/02-02-2021/Results_{shot}_train_shot_aggregated_max.csv', index=False))
```
